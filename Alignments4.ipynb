```python
from scripts.data_management import Opus100Manager
from os.path import join
from scripts.post_process import load_sents_from_file, align_sents
pairs = Opus100Manager.get_pairs()
import time

tls = ['gpt', 'deepl']
dm = Opus100Manager()

for tl in tls:
    for pair in pairs:
        s, t = pair
        mt_sents = load_sents_from_file(
            filename=f'opus-{tl}-{s}-{t}',
            folder='translations'
        )
        src_sents, _ = dm.get_sentence_pairs(s, t, num_of_sents=400)
        stime = time.time()
        _ = align_sents(
            src_sents=src_sents,
            tgt_sents=mt_sents,
            src_lang=s,
            tgt_lang=t,
            folder_path=join('src_mt_alignments', f'opus-{tl}')
        )
        etime = time.time()
        diff = etime-stime
        print(f'{tl}-{s}-{t} took {diff:.2f}s\n')
```
